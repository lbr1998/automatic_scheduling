```python
from datetime import datetime
from task import Task
from employee import Employee
from task_scheduler import TaskScheduler
from visualization import visualize_schedule
import csv

%load_ext autoreload
%autoreload 2
```

```python
def convert_to_task(row):
    """
    Convert a row of data into a Task object.

    Args:
        row (list): A list containing task data with the following format:
            [pick_up_time_str, user_id, pick_up_address, drop_off_address, _, priority]

    Returns:
        Task: An instance of the Task class with the provided data.
    """
    # Extract the pick-up time string from the row
    pick_up_time_str = row[0]

    # Extract the user ID from the row
    user_id = row[1]

    # Construct the pick-up and drop-off addresses, appending ", Munich, Germany"
    pick_up_address = row[2].strip() + ", Munich, Germany"
    drop_off_address = row[3].strip() + ", Munich, Germany"

    # Extract the priority and convert it to an integer
    priority = int(row[5])

    # Convert the pick-up time string to a datetime object for the departure time
    departure_time = datetime(2024, 8, 1, int(pick_up_time_str.split(':')[0]), int(pick_up_time_str.split(':')[1]))

    # Create a new Task object with the extracted and processed data
    task = Task(user_id, pick_up_address, drop_off_address, priority, departure_time)

    return task
```

```python
csv_file = 'TC24-SampleOrders - Orders.csv'

with open(csv_file, 'r', newline='', encoding='utf-8') as file:
    csv_reader = csv.reader(file, delimiter=',')
    next(csv_reader)

    tasks = []
    for row in csv_reader:
        task_str = convert_to_task(row)
        tasks.append(task_str)
```

```python
for task in tasks:
    print(f'User ID: {task.user_id}')
    print(f'Start Address: {task.start_location}')
    print(f'End Address: {task.end_location}')
    print(f'Priority: {task.priority}')
    print(f'Departure Time: {task.departure_time}')
    print(f'Arrival Time: {task.arrival_time}')
```
